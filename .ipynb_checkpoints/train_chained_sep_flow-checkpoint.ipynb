```python
import tensorflow as tf
# import tensorflow.contrib.eager as tfe
# tfe.enable_eager_execution()
import sys
sys.path.append(r'~/Pose Estimation Keras')
import cnn
from cnn import conv_base
from build_dataset import Dataset
import numpy as np
# from sklearn.preprocessing import normalize
from scipy import stats

from keras import layers
from keras import models
from keras import optimizers
import keras.backend as K
from keras.callbacks import TensorBoard, ModelCheckpoint, LearningRateScheduler

from time import time
import cv2
import os 
```

```python
total_frames = 5
frame_height = 224
frame_width = 224
frame_channels = 3
num_bodyparts = 6




def build_model():
    frame_tensor = conv_base.layers[0].get_input_at(0)
    resnet_output = cnn.convolution_network(frame_tensor)
    model_resnet = models.Model(inputs=[frame_tensor], outputs=[resnet_output])
    deconv_output = cnn.deconvolution_network(model_resnet.output)
    
    heatmap_generator = models.Model(inputs=[frame_tensor], outputs=[deconv_output])
    heatmap_generator_optimizer = optimizers.Adam(lr=0.0001,
                                                        beta_1=0.9,
                                                        beta_2=0.999,
                                                        epsilon=None, 
                                                        decay=0.0, 
                                                        amsgrad=True)
    heatmap_generator.compile(optimizer=heatmap_generator_optimizer, loss='binary_crossentropy')
    
    
    optical_flow_input = layers.Input(batch_shape=( None, 5, 224, 224, 6))
    optical_flow_output = cnn.optical_flow_network(optical_flow_input)
    optical_flow_corrector = models.Model(inputs=[optical_flow_input],
                                     outputs=[optical_flow_output])
    
    optical_flow_optimizer = optimizers.Adam(lr=0.0001,
                                                        beta_1=0.9,
                                                        beta_2=0.999,
                                                        epsilon=None, 
                                                        decay=0.0, 
                                                        amsgrad=True)
    optical_flow_corrector.compile(optimizer=optical_flow_optimizer, loss='mse')
    return heatmap_generator, optical_flow_corrector

heatmap_generator, optical_flow_corrector = build_model()
heatmap_generator.summary()
print('\n')
optical_flow_corrector.summary()
```

```python
import csv
def parse_csv(file):
    with open(file) as csv_file:
        csv_reader = csv.reader(csv_file, delimiter=',')
        indices = []
        for line in csv_reader:
            indices.extend([int(f) for f in line if f!=''])
    return indices

indices = parse_csv('dataset_indices.csv')
print(indices)
```

```python
import h5py
with h5py.File('../dataset.h5', 'r') as hf:
    central_heatmap = hf['central_heatmap'][indices]
    frames = hf['frames'][indices]
    heatmaps_individual = hf['heatmaps_individual'][indices]
    heatmaps_combined = hf['heatmaps_combined'][indices]
    
home_dir = os.getcwd()
central_heatmap = np.expand_dims(central_heatmap, axis=-1)
central_heatmap = np.swapaxes(central_heatmap, 1, 4)
heatmaps_combined = np.expand_dims(heatmaps_combined, axis=-1)
heatmaps_individual = np.swapaxes(heatmaps_individual, 2, 4)
print(central_heatmap.shape)
print(heatmaps_individual.shape)
print(heatmaps_combined.shape)
```

```python
def normalize(x):
    x = np.array(x, dtype = np.float64)
    x -= np.mean(x, dtype = np.float64)
    x /= np.std(x, dtype = np.float64)
    return x


def generator(frames, central_heatmap, heatmaps_individual):
    while(1):
        for j,k,l in zip(frames, central_heatmap, heatmaps_individual):
            for channel in range(l.shape[0]):
                l[channel, :, :, :] = normalize(l[channel, :, :, :]) 
            yield j,k,l 
            
```

```python

iterations = 2000
batch_size = 1


for step in range(iterations):
    frame,c_h_map,h_map = next(generator(frames,central_heatmap, heatmaps_individual))
    
    h_loss = heatmap_generator.train_on_batch(frame, h_map)
    
#     if step%3 == 0 and iterations>10000:
#         h_map = heatmap_generator.predict(frame)
        
    h_map, c_h_map = np.expand_dims(h_map, axis = 0), np.expand_dims(c_h_map, axis = 0)
    o_loss = optical_flow_corrector.train_on_batch(h_map, c_h_map)
    
    
    if step % 100 == 0:
        print(step)
        heatmap_generator.save('pretrained_models/heatmap_generator_1.h5')
        optical_flow_corrector.save('pretrained_models/optical_flow_corrector_1.h5')
        print('heatmap_generator loss:', h_loss)
        print('optical_flow_corrector loss:', o_loss)
    
    
```
